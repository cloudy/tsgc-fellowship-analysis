```python
import pandas as pd
import numpy as np
```

```python
# scrape rules
start_year = 2008
end_year = 2019
url_pre = 'http://www.tsgc.utexas.edu/grants/'
award_type = 'fellows'
all_data = pd.DataFrame()
```

```python
for year in range(start_year, end_year + 1):
    try:
        print(year)
        url = url_pre + str(year) + '/' + award_type + '.html'
        df = pd.read_html(url, header=0)
        df = df[np.argmax([len(d) for d in df])]
        df['YEAR'] = year
        all_data = all_data.append(df, sort=False, ignore_index=True)
    except:
        print('Year not found:',year)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(all_data)
```

```python
all_data.groupby('INSTITUTION').size().sort_values(ascending=False)
```

```python
uta_data = all_data[all_data['INSTITUTION'].str.contains('arlington', na=False, case=False)]
unique_uta_names = uta_data['NAME'].unique()

for name in unique_uta_names:
    match = uta_data['NAME'] == name
    
    if match.sum() >= 2:
        display(uta_data[match])
        print('won %d years!' % match.sum())
        
```
